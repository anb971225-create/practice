### Numpy 데이터 접근
- 인덱싱, 슬라이싱 : 위치(인덱스)개념 기반
- 불리언 인덱싱: 논리연산자를 활용하여 True, False 결과 출력 -> True 위치의 데이터 접근

```python
arr = np.array([10,20,30,40,50,60])

arr = np. arange(10,61,10)

arr
```

```python
arr[[3,5]]
```

```python
# 슬라이싱 20~50 데이터접근

arr[1:5]
```

```python
# 10, 30 데이터 접근 (슬라이싱)

arr[0:3:2]

arr[:3][[0,2]]
```

```python
# arr 2차원 변경
arr2 = arr.reshape(2,3)

arr2
```

```python
# 2차원 배열 접근하기
# arr2[ 행의 범위, 열의 범위 ] 

# arr[0] -> 0번째 행 접근
# arr[0,1] -> 0번째 행 , 1번째 열 접근
# ,를 통해 방향(축) 구분 가능
print(arr2[0,1])
```

```python
arr2[1,[0,1]]
```

```python
arr3 = np.arange(1,51).reshape(5,10)

print(arr3[:6, 0])
print(arr3[:6, 3])
print(arr3[:6, 6])
print(arr3[:6, 9])
```

```python
arr3[ : , [0,3,6,9]]
```

```python
arr3[ : , 0:10:3] # arr3[ : , ::3]으로도 가능
```
