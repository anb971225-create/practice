### 동영상 출력하기

-cv2.wait(밀리초) : 동영상이 실행되는 시간에 대해서 제어

```python
cap = cv2.VideoCapture('./data/dogs.mp4') # 영상(비디오 객체)가져옴

while True: # 계속 실행한다, 동영상: 연속적인 이미지
    ch, frame = cap.read() # ch: 성공여부(True or False), frame: 받아온 이미지

    #이미지를 불러올 때 실패할 경우 while 멈추기
    #ch가 False일 경우, 실행되는 break
    if not ch:
        print('동영상 재생 끝 or 읽기 실패')
        break

    # 제대로 읽을 경우
    cv2.imshow('video1', frame) # 비디오창 이름, 받아온 이미지(프레임)

    # q를 누르면 -> 동영상 윈도우창 꺼짐
    key = cv2.waitKey(1) # 속도 제어, 사용자 입력값은 key에 저장
    if key == 113:
        print('사용자에 의해서 영상이 종료되었다')
        break

cap.release() # 카메라 연결 자원 해제
cv2.destroyAllWindows()
```

- cv2.waitkeay(delay) 함수: 프레임간의 간격을 제어하고 사용자의 입력을 처리할 수 있음
  - 주어진(밀리초)동안 사용자의 키 입력을 기다림, 사용자의 입력이 들어올때까지 설정한 시간동안 대기하며 해당키의 아스키 코드를 반환
  - 매개변수 delay는 대기 시간을 나타내는 값으로 밀리초 단위
  - 1000ms는 1s 대기, 500ms는 0.5s (다음 이미지를 보여주기까지의 대기시간)

- frame 단위로 캡쳐한 이미지 저장

```python
cap = cv2.VideoCapture('./data/dogs.mp4') # 영상(비디오 객체)가져옴

cnt = 0

while True: # 계속 실행한다, 동영상: 연속적인 이미지
    ch, frame = cap.read() # ch: 성공여부(True or False), frame: 받아온 이미지

    #이미지를 불러올 때 실패할 경우 while 멈추기
    #ch가 False일 경우, 실행되는 break
    if not ch:
        print('동영상 재생 끝 or 읽기 실패')
        break

    # 제대로 읽을 경우
    cv2.imshow('video1', frame) # 비디오창 이름, 받아온 이미지(프레임)

    # q를 누르면 -> 동영상 윈도우창 꺼짐
    key = cv2.waitKey(1) # 속도 제어, 사용자 입력값은 key에 저장
    if key == 113:
        print('사용자에 의해서 영상이 종료되었다')
        break

    # 사진 찍기, 프레임 저장하기
    if key == ord('w'):
        cnt += 1 # cnt = cnt +1
        print(f'./data/dog_img{cnt}.png 사진 저장')
        cv2.imwrite(f'./data/dog_img{cnt}.png', frame, params =[cv2.IMWRITE_PNG_COMPRESSION, 0])
        # imwrite : 이미지 저장 함수, 
        # imwrite(경로, 저장할 이미지, 이미지 저장시의 옵션(압축정도, 0일경우 원본 유지, 9일경우 가장많이 압축)) 

cap.release() # 카메라 연결 자원 해제
cv2.destroyAllWindows()
```

```python
import cv2

# 카메라로부터 영상 획득

cap = cv2.VideoCapture(0)

# 녹화 설정
fps = 30.0 # 초당 프레임 이미지 수(1초에 30장)

#영상의 크기
#w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
w = int(cap.get(3))
h = int(cap.get(4))

# cap.get(ID) 3: 가로크기 , 4: 세로크기, 5 : FPS,  7: 전체 프레임 수

# 코덱 설정
codec = cv2.VideoWriter_fourcc(*'MP4V') # mp4 형식으로 저장

# 녹화 파일 설정
out = cv2.VideoWriter('./data/sample.mp4', codec, fps, (w,h))
# 경로, 코덱, 프레임 , (너비, 높이)

# 녹화 여부 설정
record = False

while cap.isOpened(): # cap이라는 객체가 잘 열려있다면,
    ch, frame = cap.read()

    if not ch:
        print('이미지 읽기 실패 또는 전체 영상 플레이 완료')
        break

    cv2.imshow('video', frame)
    key = cv2.waitKey(30)

    if key == ord('1'):
        print('녹화 종료')
        break
    elif key == ord('2'):
        print('녹화 시작')
        record = True
    # 녹화 시작
    if record:
        out.write(frame)

cap.release() #카메라 종료
out.release() #파일 저장 종료

cv2.destroyAllWindows()
```

### 저장한 이미지들을 동영상으로 만들기

```python
import glob

# 첫 이미지로 크기 가져오기

img = cv2.imread('./data/CAPTURE/1.png')
h, w, _ = img.shape

img_list = glob.glob('./data/CAPTURE/*.png')

fps = 10.0

codec = cv2.VideoWriter_fourcc(*'DIVX') # avi파일 코덱

out = cv2.VideoWriter('./data/sample2.avi', codec, fps, (w,h))

for i in range(len(img_list)): # 캡쳐한 사진 수만큼 반복
    # 저장된 이미지들을 읽어온다
    img = cv2.imread(f'./data/CAPTURE/{i+1}.png',cv2.IMREAD_COLOR)
    out.write(img) # 영상 저장


out.release()
```
