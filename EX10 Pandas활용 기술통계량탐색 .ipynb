## 기술통계 확인하기(요약, 수치)
- 기술 통계 확인하는 이유
  - 데이터 분석에서 가장 중요한 첫 단계는 데이터를 이해하는 것
  - 데이터 범위·대표값·이상치 파악
  - 데이터 비율·경향·불균형 파악
  -> 결측치가 생기면 단순 삭제하는것이 아니라 기술통계량을 확인하여 대체할만한 값들을 탐색하여 결측치 처리

### data.describe()
- 숫자형 데이터의 전체적인 ‘요약 정보’를 빠르게 파악할 수 있음

- 반환되는 정보

  - count : 데이터 개수
  - mean : 평균
  - std : 표준편차 -> 데이터의 분포 파악 -> 분산에 루트를 씌운 값, 데이터가 얼마나 흩어져 있는가
  - min, max : 최소값, 최댓값
  - 25%, 50%(중위수, 중앙값), 75% : 사분위수(데이터의 분포를 4등분 확인)

### data.describe(include = 'object')

- 문자형(object, 범주형) 데이터만 선택해 기술통계를 보여줌

- 반환되는 정보

 - count : 데이터 개수

 - unique : 고유값 개수

 - top : 가장 많이 등장한 값

 - freq : top 값의 등장 빈도

```python
edu_uni = hr_info['EDU_LEVEL'].unique()

edu_uni
```

```python
#박사 학위의 평균 연봉 구하기
#1. 박사학위 데려오기

a = hr_info[hr_info['EDU_LEVEL'] == '박사 학위'][['SALARY']]
A = hr_info[hr_info['EDU_LEVEL'] == '학사 학위'][['SALARY']]
b = hr_info[hr_info['EDU_LEVEL'] == '석사 학위'][['SALARY']]
c = hr_info[hr_info['EDU_LEVEL'] == '고등학교 졸업'][['SALARY']]
d = hr_info[hr_info['EDU_LEVEL'] == '전문대 학위'][['SALARY']]

print(a['SALARY'].mean().round(2))
print(A['SALARY'].mean().round(2))
print(b['SALARY'].mean())
print(c['SALARY'].mean())
print(d['SALARY'].mean())
```

```python
print(hr_info[hr_info['EDU_LEVEL'] == '박사 학위']['SALARY'].mean().round(2))
```

```python
for i in edu_uni:
    result = hr_info[hr_info['EDU_LEVEL'] == i]['SALARY'].mean().round(2)
    print(f"{i}: {result}")

#Pandas에서 group by는 위와같은 for i 함수를 도와주는 도구다.
```

- 데이터 그룹화를 이용한 통계분석
  - groupby(): 카테고리별 그룹을 만들어 연산 수행

```python
# 교육 수준에 따른 평균 연봉 구하기
# 데이터.groupby(기존 컬럼명).연산함수
# EDU_LEVEL, SALARY
hr_info[['EDU_LEVEL', 'SALARY']].groupby('EDU_LEVEL').mean().round(2)

# 카테고리로 설정할 데이터가 들어있는 컬럼을 기준컬럼이라고 한다! 
```

```python
hr_info.columns
```

```python
# 교육수준과 성별로 나눠서 평균연봉 확인하기

hr_info[['EDU_LEVEL','SALARY','GENDER']].groupby(['EDU_LEVEL','GENDER']).mean().round(2)
```

```python
# 성별에 따른 평균연봉 구하기

hr_info[['GENDER','SALARY']].groupby('GENDER').mean().round(2)
```

- aggregate(집계하다, 요약하다)
- agg : 요약 통계를 한 번에 여러 개 뽑고 싶을 때 사용, 즉 agg()는 그룹별로 묶은 데이터를
어떤 방식으로 요약할지 지정하는 함수

```python
hr_info[['GENDER','SALARY']].groupby('GENDER').agg(['mean','max','min','median']).round(2) #agg는 그룹바이함수에서만 사용한다.
```
