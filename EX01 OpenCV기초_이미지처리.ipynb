## 이진(Binary) 이미지 구성(이미지 이진화)
- 0,1 데이터 구성
- cv2.threshold(img, threshold_value, value, flag)
- img : Grayscale 이미지
- threshold_value : 픽셀 임계값(0과 1로 나누기 위한 기준값)
- value : 임계값보다 클 때 적용되는 값
- flag : 임계값 적용방법
  - cv2. THRESH_BINARY : 임계값보다 크면 value, 아니면 0을 할당
  - cv2. THRESH_BINARY_INV : 임계값보다 크면 0, 아니면 value를 할당
  - cv2. THRESH_TRUNC : 임계값보다 크면 value, 아니면 픽셀값은 변경되지 않음

```python
# 흑백 이미지
img = cv2.imread('./data/ryan.jpg',cv2.IMREAD_GRAYSCALE)

# 이미지 이진화
_, thr_img = cv2.threshold(img, 100,255,cv2.THRESH_BINARY)

# 오츠 이진화

th_val, thr_img2 = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

#0은 의미없는 값이다

# plt.subplot(행, 열, 순서)

plt.subplot(1,3,1)

plt.title('original')
plt.axis('off')

plt.imshow(img, cmap= 'gray')

plt.subplot(1,3,2)

plt.title(f'binary:{_}')
plt.axis('off')

plt.imshow(thr_img, cmap= 'gray')

plt.subplot(1,3,3)

plt.title(f'otsu binary:{th_val}')
plt.axis('off')

plt.imshow(thr_img2, cmap= 'gray')
```

```python
# 컬러로 가져온 후 흑백사진으로 변환

apple = cv2.imread('./data/apple.jpg',cv2.IMREAD_COLOR)

apple_gray = cv2.cvtColor(apple, cv2.COLOR_BGR2GRAY)

height, width = apple_gray.shape
center_x, center_y = width//2, height//2

print(height, width)
print(center_x, center_y)

# 가운데 정사각형 영역 접근
# 좌상단 좌표

top_left_x, top_left_y = center_x//2, center_y//2
print(top_left_x, top_left_y)

bottom_right_x = center_x + top_left_x 
bottom_right_y = center_y + top_left_y

print(bottom_right_x, bottom_right_y)
# 배열[ 행의범위, 열의범위]



# 사각형 영역 추출하기
rect_region = apple_gray[top_left_y:bottom_right_y,top_left_x :bottom_right_x]
plt.imshow(rect_region,cmap='gray')
plt.axis('off') 

plt.show()
```

```python
# rect_region 이진화 변형 - otsu

val, apple_bin = cv2.threshold(rect_region, 0 , 255 , cv2.THRESH_BINARY | cv2.THRESH_OTSU)

plt.axis('off')
plt.imshow(apple_bin, cmap = 'gray')
```

```python
# 원본 이미지에 이진화 변형 이미지 대입하기

# apple2 = apple_gray #얕은 복사( 주소 복사, 원본(apple_gray)의 내용이 바뀜)
apple2 = apple_gray.copy() # 깊은 복사 (원본 형태  복사, 원본(apple_gray)의 내용이 바뀌지않음)

apple2[top_left_y:bottom_right_y,top_left_x :bottom_right_x] = apple_bin

plt.axis('off')
plt.imshow(apple2,cmap = 'gray')

plt.show()
```

```python
# 이미지 2개 선택해서 불러오기

img1 = cv2.imread('./data/ryan.jpg')
img2 = cv2.imread('./data/apple.jpg')

# 색상공간 변환 bgr -> rgb

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# 이미지 크기 변경 - 사이즈가 동일해야 블렌딩 가능

img1 = cv2.resize(img1, dsize = (300,300))
img2 = cv2.resize(img2, dsize = (300,300))


# cv2.addWeighted(이미지1, 가중치, 이미지2, 가중치 , 0(전체가중치))

result = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)

# 출력 

# 위치잡기
plt.subplot(1,3,1)
# 그림 넣기
plt.imshow(img1)

plt.subplot(1,3,2)
# 그림 넣기
plt.imshow(img2)

plt.subplot(1,3,3)
plt.imshow(result)
```

```python
####### q 누르면 영상 종료
import cv2
cap = cv2.VideoCapture(0)

while True:
    ch, frame = cap.read()
    if not ch:
        print('동영상 재생 끝 or 읽기 실패')
        break

    cv2.imshow('video1', frame)
    
    key = cv2.waitKey(33) 
    if key == ord('q'):
        print('사용자에 의해서 영상이 종료')
        break
cap.release()
cv2.destroyAllWindows()
```
