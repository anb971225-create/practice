## BMI 구하기 실습
- 파일 불러오는 방법 알기
- numpy 배열 연산하기!

```python
# txt 파일 불러오는 방법!
# np.loadtxt('파일의 경로'.확장자 ,데이터 구분자)
# ./ : 현재경로(생략가능!)
# delimiter : 구분자 txt 파일에서 numpy 배열의 요소로 구분해서 넣어줌
data = np.loadtxt('./data/height_weight.txt', delimiter=',')

print(data)

print(f'모양 : {data.shape}')
print(f'차원 : {data.ndim}')
print(f'타입 : {data.dtype}')
print(f'개수 : {data.size}')

print(f'크기 : {len(data)}')
```

```python
# 원본 데이터(data) 
# 1. 키 데이터 변수에 담아주기
h = data[0]

h

# 2. 몸무게 데이터 변수 담아주기

w = data[1]

w

# 3. 키 데이터의 단위를 변환!(cm -> m)


h = h * 0.01
h

# 4. BMI 지수 구하기

bmi = w / h**2

bmi
```

```python
# 5. 과체중과 비만인 데이터 찾기 -> boolean 색인 사용!

bmi >= 23 # 참인지 거짓인지만을 알 수 있음!

# 실제로  True에 해당하는 데이터를 가져오자!

bmi[bmi >= 23]
```

```python
# 6. 과체중에 해당하는 데이터만 가져오기! 23 이상 25미만

bmi[(bmi >=23) & ( bmi <25)]  # and는 사용하지못한다.
```
