## 딕셔너리(dictionary)
- 말 그대로 사전! key값과 value값을 한 쌍으로 갖는 데이터 타입
- immutable(불변하는) key와 mutable(변할수있는) value로 묶여있어 Key값으로 value값을 가져오기 때문에 순서가 없는 집합!
- 딕셔너리 만드는 법 -> 딕서너리명 = {key1:value1, key2:value2, ....}

```python
a = {}
type(a)
```

```python
list1 = [1,2,3]
tuple1 = (1,2,3)
type(list1)
```

```python
dic1 = {"name":"JUHUI","age":20}
dic1
```

```python
# 딕셔너리에 새로운 값을 추가하는 방법!
# []에 추가할 key값을 적어준다!
dic1["phone"] = "010-3228-5850"

dic1
```

```python
# 딕셔너리의 값(value)을 가져오는 방법
# []안에 key값을 적어서 가져온다!

dic1["age"]
```

```python
# .get(key값)

phone = dic1.get("phone")
print(phone)
```

```python
dic2 = {1:1,2:2,3:3}
dic2
```

```python
dic2[2]
```

## 딕셔너리 삭제
- del 키워드 사용

```python
dic2= {1:1, 2:2 , 3:3}
```

```python
# del 키워드 사용 방법
# del 딕셔너리명[key값]

del dic2[2]

print(dic2)
```

```python
## 딕셔너리의 모든 key값과 value값을 확인하는 방법
- key값만 보는 방법 -> 딕셔너리.keys()
- value값만 보는 방법 -> 딕셔너리.values()
```

```python
dic1
```

```python
#1. 딕셔너리의 key값 확인
#2. 딕셔너리의 value값 확인
```

```python
dic1.keys()
```

```python
dic1.values()
```

```python
# for문을 활용해서 모든 key값 출력하기
# for문을 활용해서 모든 value값 출력하기
```

```python
for i in dic1.keys():
    print(i, end=" ")
```

```python
for i in dic1.values():
    print(i, end=" ")
```

```python
# key값이 존재하는지 확인하는 방법
dic1
```

```python
# in 키워드 활용!
"name" in dic1
```

```python
# 딕셔너리 내부의 모든 값을 삭제하는 기능!
# 딕셔너리.clear()

dic1.clear()
```
