## 반복문
- 프로그램 내에서 똑같은(비슷한) 명령을 일정 횟수만큼 반복시켜 수행할 때 사용하는 문법
- 종류 : while문(주로 반복 횟수가 정해지지 않았을때), for문(주로 반복 횟수가 정해져 있을때)
   

```python
while 조건식 :
    실행문장1
    실행문장2
```

```python
# while문을 사용해서 1 2 3 출력하기

num =1 

while num < 4:
    print(num)
    num += 1
```

```python
# while문을 사용해서 "파이썬 최고!" 13번 출력하기!
```

```python
a = 1

while a < 14:
    print("파이썬 최고!")
    a += 1
```

```python
while True:
    print("무한반복")
    break
```

```python
# while문의 조건식을 True로 하고 
# while문 안쪽에서 조건식을 사용해
# 해당 조건을 만족하면 break 키워드 사용해서 탈출하기!

num = 1

while True:
    print(num)
    num += 1
    if num == 4:
        break
```

```python
a = int(input("현재 몸무게 :"))
b = int(input("목표 몸무게 :"))
c=1

while True:
    d = int(input(f"{c}주차 감량 몸무게 : "))
    a -= d
    c += 1
    if a < b:
        print(f"{a} kg 달성!! 축하드립니다")
        break
```

```python
pre = int(input("현재 몸무게 : "))
goal = int(input("목표 몸무게 : "))

cnt = 1

while True:
    wei = int(input(f"{cnt}주차 감량 몸무게 : "))
    # 현재 몸무게 - 감량 몸무게 -> 현재 몸무게
    pre -= wei 
    cnt += 1
    if pre < goal:
        print(f"{pre}kg 달성!! 축하합니다!")
        break
```

```python
# range() : 시작할 숫자, 종료할 숫자+1 , 증가(감)량 )
```

```python
#range(시작값, 종료값+1, 증감량)
for i in range(1,10,1):
    print(i)
```

```python
# print() 함수의 end 속성 사용하기 : 가로로 출력
for i in range(1,10,1):
    print(i, end=" ")
```

```python
for i in range(1,10,1):
    print(i, end="\t")
```
