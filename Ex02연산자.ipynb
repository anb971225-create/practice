## 논리연산자
- and, or, not

```python
# and연산자(곱 연산자) : 두 조건이 모두 참(True)여야 결과값이 True
print(3>5 and 10==20)
print(3<5 and 10==20)
print(3<5 and 10!=20)
```

```python
# or연산자(합 연산자) : 하나라도 True면 전체 결과값이 True

print(3>5 or 10==20)
print(3<5 or 10==20)
print(3<5 or 10!=20)
```

```python
# not연산자(부정연산자) : 결과값을 반대로 반환해주는 역할

print(3>5)
print(not 3>5)
```

## 삼항연산자
- (True일때 실행문) if (조건식) else (False일때 실행문)

```python
score = 50

print("합격") if score >= 60 else print("불합격") 
```
