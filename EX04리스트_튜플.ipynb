## 리스트
- 파이썬의 자료구조중 하나
- 순서가 있다! -> index 값을 가지고 있다!
- 수정이 가능하다!
- 대괄호[]로 작성되며, 콤마(,)로 구분한다!
- 인덱스를 활용해서 인덱싱, 슬라이싱 활용 가능!
- 데이터에 접근하는 방법 -> 리스트명[인덱스번호]

```python
list1 = [2, 5, 7, 9, 10]

print(list1[0], list1[3])
print(list1[2]+list1[-1])
```

```python
list2 = [1, 2, 3, ['a','b','c']]

print(list2[3]) #list2[3] == ['a','b','c']
print(list2[3][1])
```

```python
list3 = [0, 1 ,2 ,3 ,4]
print(list3[1:3])
print(list3[:2])
print(list3[-2:])
```

```python
# 리스트 더하기
list4 = [1, 2, 3]
list5 = [3, 4, 5, 6]

print(list4 + list5)
```

```python
# 리스트에 값 추가하기
# -> 리스트명.append(값) : 리스트의 맨 마지막에 삽입
# -> 리스트명.insert(인덱스, 값): 원하는 인덱스에 삽입

list6 = [0, 1, 2, 3, 4]

list6.append(5)

list6
```

```python
list6.insert(1, 6)

list6
```

```python
list6.insert(1, ['a','b'])
list6
```

```python
list6.insert(3, [7, 8, [9, 10]])

list6
```

```python
print(list6[3][2][1])
```

```python
list7 = [0, 1, 2, 3]
print("수정 전:", list7[1])
list7[1] = 7
print("수정 후:",list7[1])
```

```python
# 값 삭제하기
# -> del 리스틈명[인덱스번호] -> 없는 인덱스인 경우 에러!
# -> 리스트명.remove(값) -> 값이 없을 경우 에러!

list8 = [0, 1, 2, 3, 4, 5]
del list8[5]
list8

del list8[2:]
list8
```

```python
list9 = ['a','b','c','d','e']

list9.remove('b')
list9
```

```python
# 리스트 정렬하기
list10 = [7, 99, 13, 100, 3]

# .sort() : 오름차순으로 정렬
list10.sort()
list10
```

```python
list10 = [7, 99, 13, 100, 3]

# .reverse() : 데이터를 역순으로 재배치

list10.reverse()

list10
```

```python
list10 = [7, 99, 13, 100, 3]

# sort(), reverse() 함수 사용해서 내림차순으로 배치

list10.sort()

list10.reverse()

list10
```

```python
list10 = [7, 99, 13, 100, 3]

# sort(reverse=True) -> 내림차순 배치
list10.sort(reverse=True)
list10
```

```python
# 리스트 안에서 내가 찾고싶은 값이 몇 번째 인덱스에 있는지 아는 방법
# 리스트명.index(값)

list11 = ['a','b','c','d']

list11.index('c')
```

```python
# pop() : 값을 지우는 함수
# -> del, remove()와 달리 지운 값을 돌려줌
# pop() : 리스트의 맨 마지막 값 삭제
# pop(인덱스번호) : 원하는 인덱스의 값 삭제

list12 = ['a','b','c']

list12.pop()
```

```python
# 리스트의 길이를 알 수 있는 함수
list13 = [1, 2 , 3]
# len(리스트)
len(list13)
```
