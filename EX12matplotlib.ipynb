### 시각화 도구
- 시각화라이브러리
- 한땀한땀 사용자화 하여 그래프를 출력하거나 커스터마이징 하는것에 용이

```python
# 1. 라이브러리 불러오기!

import numpy as np # 수치 계산용 라이브러리
import pandas as pd # 행과 열을 가지는 표와같은 데이터를 다루는 라이브러리
import matplotlib.pyplot as plt # 시각화 라이브러리(색감다양, 다양한 그래프 포함)
```

```python
# 2. 데이터 불러오기
hr_info = pd.read_csv('./data/hr-info.csv',encoding = 'EUC-KR')
```

- 학위별 비율 시각화
- pie 차트
  

```python
# 주피터 노트북에서 그래프에 한글이 깨지지 않도록 글꼴(폰트) 설정
from matplotlib import rc
# 그래프에서 사용할 글꼴을 'Malgun Gothic(맑은 고딕)'으로 지정
rc('font', family = 'Malgun Gothic')
```

```python
# 학력('EDU_LEVEL')별 인원 수를 계산한 뒤 원형 그래프로 시각화

hr_info['EDU_LEVEL'].value_counts().plot(
    kind = 'pie', #바는 bar 세로바는 barh
    autopct = '%.1f%%', #각 조각의 비율을 자동으로 써주는 옵션
    title = '학위별 비율', # 그래프 제목 설정
    colormap = 'Dark2', #그래프 색상 테마 설정
    counterclock = False # 원형그래프를 시계방향으로 표시해주는 역할
)

# 멧플러립은 그래프 출력할때
plt.show()
```

```python
# plt.plot(데이터 셋)
```

```python
# 압사일 데이터가 문자열로 되어 있습니다. -> 날짜 타입으로 변경
hr_info['HIRE_DATE'] = pd.to_datetime(hr_info['HIRE_DATE'])
hr_info['HIRE_YEAR'] = hr_info['HIRE_DATE'].dt.year
hr_info['HIRE_YEAR'].value_counts().sort_index()
hire_year_rs = hr_info['HIRE_YEAR'].value_counts().sort_index()
```

```python
plt.plot(hire_year_rs)
```

```python
plt.figure(figsize = (10,4))  # figure : 전체도화지(큰틀)
plt.plot(
    hire_year_rs,
    ls = ':',
    lw = 20,
    color = '#F2FF01FF', #선의색, 헥스코드
    marker = '$😀$',
    mfc = 'black',
    mec = 'pink',
    ms = 10,
    mew = 1,
    label = '인원수'
)

plt.xlabel('입사년도')
plt.title('연도별 입사자 수')

plt.show
```
