## 외장함수(외부 라이브러리)
- 파이썬 표준 라이브러리 외에 추가적으로 설치하여 사용할 수 있는 라이브러리나 모듈에서 제공하는 함수를 의미
- 특정 작업을 더 간단하고 쉽게 수행할 수 있거나, 새로운 기능을 사용할 수 있게 해주는 함수

## 사용자 정의 함수
- 실제로 사용자가 필요한 기능을 만드는 함수

```python
# 계산기 함수 만들기
# num1(정수), num2(정수), op(문자)
```

```python
def cal(num1, num2, op):
    answer = 0
    # 더하는 기능
    if op == "+":
        answer = num1 + num2
    #빼기 기능
    elif op == "-":
        answer = num1 - num2
    #곱하기 기능
    elif op == "*":
        answer = num1 * num2
    # 나누기 기능
    else:
        answer = num1 / num2  
    return answer
```

```python
print(cal(4,6,"*"))
cal(5,8,"/")
```

- print : 단순히 값을 출력하기만 하고 외부에서 사용하지 않아도 될 때 사용
- return은 함수를 더 유연하게 재사용 가능하게 만들어 주기 때문에 결과값을 반환할 때, 또는 결과를 다른 곳에 사용할 때 return키워드를 사용하고, 즉시 값을 출력하기만 하고 따로 결과값을 저장할 필요가 없을 때 print를 사용한다.

```python
# 독스트링(docstring) : 함수의 설명을 작성 -> Shift + Tab

def cal(num1, num2, op):
    """덧셈과 뺄셈을 계산하는 함수"""
    if op == "+":
        return num1 + num2
    else:
        return num1 - num2
```

```python
cal(1,2,"+")
```

```python
# 함수의 매개변수를 기본값으로 설정하는 방법
# 2^3 -> 2**3

def power_of_n(num, power=2):   # power =2 는 디폴트값, 임의값을 집어넣으면 임의의값으로 계산해준다.
    return num**power
```

```python
power_of_n(5)
```
