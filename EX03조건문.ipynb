## 조건문
- 조건에 따라서 실행하는 문장이 달라짐
- 종료 : if(만약에), elif(else if)(그 이외에), else(아니라면)

```python
if 조건식:
    실행문장1
    실행문장2
실행문장3
```

```python
if True:
    print("실행문장")
```

```python
if False:
    print("실행문장")
```

```python
money = 11000

if money >= 10000:
    print("택시를 탄다")
print("if문 탈출!")
```

```python
if 조건식:
    실행문장1
else:
    실행문장2
```

```python
if False:
    print("실행문장1")
else:
    print("실행문장2")
```

```python
# money가 10000원 이상이라면 "택시를 탄다!" 출력
# money가 10000원 미만이라면 "버스를 탄다!" 출력
```

```python
money = 10000

if money >= 10000:
 print("택시를 탄다!")
else:
 print("버스를 탄다!")
```

```python
num= int(input("정수입력 >> "))

if num%3==0 and num%5 ==0:
    print("3과 5의 배수입니다")
else:
    print("3과 5의 배수가 아닙니다")
```

```python
num = int(input("정수 입력>>"))

if num%3 == 0 and num%5 == 0:
    print("3과 5의 배수입니다")
else:
    print("3과 5의 배수가 아닙니다")
```

```python
num = int(input("사려는 상품의 갯수를 입력하세요 >>"))

if num >= 11:
    num = num - num/10

print(f"가격은 {int(num*10000)}원 입니다")
```

```python
cnt = int(input("사려는 상품의 갯수를 입력하세요 >>"))

if cnt > 10:
    # 10% 할인 적용
    print(f"가격은 {cnt*10000*0.9}원 입니다")
else:
    print(f"가격은 {cnt*10000}원 입니다")
```

```python
num1 = int(input("첫 번째 정수 입력 >> "))
num2 = int(input("두 번째 정수 입력 >> "))

if num1 > num2:
    print("첫 번째 정수가 더 큽니다.")

elif num1 < num2:
    print("두 번째 정수가 더 큽니다.")
else:
    print("두 수가 똑같습니다.")
```

```python
score = int(input("점수 입력 >> "))

if score >= 90:
    print(f"{score}점은 A학점입니다.")
elif score >= 80:
    print(f"{score}점은 B학점입니다.")
elif score >= 70:
    print(f"{score}점은 C학점입니다.")
elif score >= 60:
    print(f"{score}점은 D학점입니다.")
else:
    print(f"{score}점은 F학점입니다.")
```

```python
score = int(input("점수 입력 >> "))

if score >= 90:
    grade="A"
elif score >= 80:
    grade="B"
elif score >= 70:
    grade="C"
elif score >= 60:
    grade="D"
else:
    grade="F"

print(f"{score}점은 {grade}학점 입니다.")
```
