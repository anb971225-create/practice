```python
# 문자열과 관련된 기술 
# 1. 인덱싱(index-> 위치)
# -> 무언가를 가르키다
# -> 문자열, 자료구조(배열, 리스트)에서 특정한 값 하나를 꺼내올 때 사용

name = "My name is JH"
# M -> 0번 인덱스에 있다

# 인덱싱, 슬라이싱을 활용할 때는 변수 뒤에 []를 사용해서 꺼내올 인덱스의 번호를 적는다!!
name[0]
# 음수 인덱싱도 가능하다!
name[-2]
```

```python
print(name[-2]+name[-1])
```

```python
# 슬라이싱
# -> 자료를 여러개 가져올 때 사용하는 기술
# -> 문자열, 자료구조 안에서 특정 범위만큼의 자료를 잘라온다
name = "My name is JH"

print(name[0:2])
```

```python
name = 'My name is JH'
print(name[0]+name[1])
print(name[3:7])
print(name[11:13])
```

```python
name = 'My name is JH'

# is JH

print(name[8:])
print(name[-5:])
print(name[:])
```

### 포매팅
- 문자열 내부에 내가 원하는 위치에 원하는 값을 변경하거나 삽입할 때 사용
- 변수 내부의 값을 바꿀 때 많이 사용

```python
# 1. %포매팅
# %d(정수), %s(문자열)

# 10월 25일의 점심은 토스트다!
month = 10 
day = 25
lunch = "토스트"

day_lunch = "%d월 %d일의 점심은 %s다!"%(month, day, lunch)
day_lunch
```

```python
# 2. format 함수를 통한 포매팅
# 제 이름은 이주희이고, 취미는 코딩하기입니다~

name = "이주희"
hobby = "코딩하기"

my_hobby = "제 이름은 {}이고, 취미는 {}입니다~".format(name, hobby)
my_hobby
```

```python
x = 100
y = 200
add = print(x + y)
result = "%d와 %d의 합은 %s입니다"%(x, y, add)
result
```

```python
x = 100
y = 200
add = x + y
result = "%d와 %d의 합은 %s입니다"%(x, y , add)
print(result)
```

```python
result = "{}와 {}의 합은 {}입니다".format(x ,y , add)
result
```

```python
# 3. f-string 포매팅
# format 함수는 2.xx버전에서 쓰던 방식
# f-string 포매팅은 3.xx버전 이후로 출시된 방식

dinner = "치킨"

print(f"어제 먹은 저녁은 {dinner}이였습니다")
```

```python
morning = "토스트"
lunch = "햄버거"
dinner = "치킨"

print(f"어제 먹은 아침은 {morning}이고 점심은 {lunch}이고 저녁은 {dinner}이다")
```

## 문자열 함수 실습

```python
s = "문자열 함수 실습 어렵지 않아요!"
```

```python
# count('문자') -> 문자열에 포함된 갯수 반환
s.count(' ')
```

```python
# find('문자') -> 문자의 위치(index)를 반환
s.find('함') # 만약에 없는 값이면 -1을 반환
```

```python
# join(변수명) -> 각 문자 사이사이마다 넣고 싶은 문자가 있을 때 사용
" ".join(s)
```

```python
# upper() -> 영문자를 다 대문자로 변환 
s1 = "hello world!"
s1.upper()

# lower() -> 영문자를 다 소문자로 변환
```

```python
s2 = "    hi    "
# lstrip() -> 왼쪽 공백 없애기
# rstrip() -> 오른쪽 공백 없애기
# strip() -> 양쪽 공백 없애기

s2.lstrip()
```

```python
s2.rstrip()
```

```python
s2.strip()
```

```python
# replace() -> 데이터를 다른 값으로 변환하고 싶을 때 사용
s3 = "pithon"

s3.replace("i", "y")
```

```python
# split("문자")
s4 = "햄버거,피자,김밥,비빔밥"
s4.split(",")
```
